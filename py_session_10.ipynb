```python
# create 4 individual functions
# fun1: add
# fun2: sub
# fun3: mul
# fun4: divi

# create a main function name: calculator
# inside main function
# print some statements
# option1=addition,
# option2=sub, opt3= mul , opt4=diivision
# option=eval(input("1:4")
# a value
# b value
# if option==1:
#   add function
# elif option==2:
#   sub function
# elif option==3
#    mul funtion
#  if option==4:   
#  div function

def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mult(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
def calculator():
    print('option-1:add')
    print('option-2:sub')
    print('option-3:mult')
    print('option-4:div')
    option=eval(input("enter options 1 to 4 : "))
    a=eval(input("enter a value:"))
    b=eval(input("enter b value:"))
    
    if option==1:
        add(a,b)
    elif option==2:
        sub(a,b)
    elif option==3:
        mult(a,b)
    elif option==4:
        div(a,b)
    else:
        print("enter the valid option 1 to 4")

calculator()
```
